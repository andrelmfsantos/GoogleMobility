```python
import pandas as pd
```

```python
# Base de dados
df = pd.read_csv("2020_BR_Region_Mobility_Report.csv")
df
```

```python
# colunas no df
df.columns
```

```python
# subset com colunas de interesse
di = df[['date','retail_and_recreation_percent_change_from_baseline', 'grocery_and_pharmacy_percent_change_from_baseline',
         'parks_percent_change_from_baseline','transit_stations_percent_change_from_baseline',
         'workplaces_percent_change_from_baseline','residential_percent_change_from_baseline']]
di
```

```python
# renomear colunas
di.columns = ['date','retail','grocery','parks','transit','workplaces','residential']
di
```

```python
# Criar coluna "month"
di['date'] = pd.to_datetime(df['date']) # converte srt para date type
di['month'] = di['date'].dt.month       # cria coluna com nº do mês
di
```

```python
# Agrupar por mês
dm = di[['month','retail','grocery','parks','transit','workplaces','residential']]
dm
```

```python
dm.groupby('month').agg('median')
```

```python
dm.to_csv("isolamento2020")
```
